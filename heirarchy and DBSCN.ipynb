```python
#import all the dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings("ignore")


from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=load_iris()
```

```python
df
```

```python
x=df.data
y=df.target
```

```python
x
```

```python
y
```

```python
#scatter plot for visual
plt.scatter(x[:,0],x[:,1])
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
#scaling the data
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
#pca decompostion for selecting relevant features
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(x)
```

```python
#let's make dendogram
```

```python
import scipy.cluster.hierarchy as sc 
```

```python
plt.figure(figsize=(15,50))
plt.title("dendrogram")
sc.dendrogram(sc.linkage(pca_scaled, method="ward"))
plt.xlabel("sample_data")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aglo=AgglomerativeClustering(n_clusters=3,metric="euclidean",linkage="ward")
```

```python
aglo.fit(pca_scaled
        )
```

```python
y_labels=aglo.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=y_labels)
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_x=scaler.fit_transform(x)
```

```python
scaled_x
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.5)
```

```python
db.fit_predict(scaled_x)
```

```python
y_label=db.labels_
```

```python
plt.scatter(scaled_x[:,0],scaled_x[:,1],c=y_label)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# generate dataset
X, _ = make_moons(n_samples=300, noise=0.1)

# fit DBSCAN
db = DBSCAN(eps=0.3, min_samples=5)
labels = db.fit_predict(X)

# scatter plot of the clusters
plt.scatter(X[:,0], X[:,1],c=labels)
plt.title("DBSCAN on make_moons")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
